## Exploratory Data Analysis

First, let's try to get a better overview of the data. We create a table and filter for no-shows only. 

```python
#Filter for No-shows only 
df_noshow = df[df['No-show'] == 'Yes']
df_noshow.reset_index(inplace = True, drop = True)

print("There were {} no-shows".format(df_noshow.shape[0]))
df_noshow.head()
```

We can create histograms to quickly compare frequencies of the characteristics. 

```python
#For all appointments: create histograms for numeric features, filtering out AppointmentID and PatientID
df[df.columns[~df.columns.isin(['AppointmentID','PatientId'])]].hist(figsize = (10,10), sharey = True);
plt.suptitle('Histograms - All Appointments', ha='center', fontsize='x-large');
plt.xlabel("Number of Patients")
```

```python
#For no-shows: create histograms for numeric features, filtering out AppointmentID and PatientID
df_noshow[df_noshow.columns[~df_noshow.columns.isin(['AppointmentID','PatientId'])]].hist(figsize = (10,10), sharey = True);
plt.suptitle('Histograms - No-Show Appointments', ha='center', fontsize='x-large');
plt.xlabel("Number of Patients")
```

Based on this preliminary inspection, it seems that the characteristics that differ most between the two groups (total appointments vs. no-shows) are "Age" and "SMS Received". We will examine those in later sections. First, we study the neighborhoods involved. 

<a id='question1'></a>
### Research Question 1: Which neighborhoods had the most and the least no-shows? 
We determine this by taking the proportion of no-shows for each neighborhood ( = number of no shows / total number of appointments ).

To count the total number of appointments per neighborhood: 

```python
counts_total = df.groupby('Neighborhood').count()
counts_total = counts_total[[ 'Gender']] #can choose any column 
counts_total.rename(columns = {'Gender': 'Total_appts'}, inplace = True) #rename column 
counts_total.head()
```

Next, we count the number of no-shows per neighborhood.

```python
counts_ns = df_noshow.groupby('Neighborhood').count()
counts_ns= counts_ns[['Gender']] #can choose any column 
counts_ns.rename(columns = {'Gender': 'Num_no_shows'}, inplace = True) #rename column 
counts_ns.head()
```

Now, we combine the two tables and calculate the proportion of appointments missed per neighborhood.

```python
neighborhoods = counts_ns.join(counts_total, how = "inner") 
neighborhoods['Prop_no_shows'] = neighborhoods['Num_no_shows']/neighborhoods['Total_appts']
neighborhoods.sort_values(by = 'Prop_no_shows', ascending = False, inplace = True)
neighborhoods.head(10)
```

For some of the neighborhoods, like the *ILHAS OCEÂNICAS DE TRINDADE* above, there are too few appointments for a robust analysis. Let's try filter out neighborhoods that have less than 50 total appointments. 

```python
neighborhoods_filtered = neighborhoods.query("Total_appts > 50")
neighborhoods_filtered.head()
```

To get the 10 neighborhoods with the *greatest* proportion of no-shows: 

```python
neighborhoods_filtered.head(10)
```

To get the 10 neighborhoods with the *least* no-shows: 

```python
neighborhoods_filtered.tail(10)
```

<a id='question2'></a>
### Research Question 2 : Which age group and gender is more likely to miss an appointment? 

In this section, we investigate which age group and gender has historically been more likely to miss an appointment. 

First, we examine the ages of all patients. We will create distinct age groups based on the quartiles. 

```python
df.Age.describe()
```

Based on the distribution of ages, we set our age group categories to be:
* **Group 1**: Ages 0 - 18 
* **Group 2**: Ages 19 - 37
* **Group 3**: Ages 37 - 55
* **Group 4**: Ages 55+ 

We create a column for "Age Group", using the pandas **.cut( )** function. 

```python
# Bin edges that will be used to "cut" the data into groups
bin_edges = [-1, 18, 37, 55, 115]

# Labels for the four age groups
bin_names = ['Group 1','Group 2','Group 3','Group 4' ] 

# Creates age_group column
df['Age_Group'] = pd.cut(df['Age'], bin_edges, labels=bin_names)

# Checks for successful creation of this column
df.head()
```

```python
df_noshow.head()
```

We can now calculate the proportion of no-shows for each age group. 

```python
#Add Age_Group column to no_show table
df_noshow_ages = df_noshow.merge(df[['AppointmentID', 'Age_Group']], left_on = 'AppointmentID', right_on = 'AppointmentID', how = 'inner')
print(df_noshow_ages.shape)
df_noshow_ages.head(10)
```

```python
#Create table containing # of total appts, # of no-shows, and proportion of no-shows for each age group
df_age = df.pivot_table(index = ["Age_Group"], values = 'AppointmentID', aggfunc = 'count')
df_age.rename(columns = {"AppointmentID": "Total_Appts"}, inplace = True)

df_age['Num_No_Shows'] = df_noshow_ages.groupby(["Age_Group"]).count()['AppointmentID']
df_age["Prop_NoShows"] = df_age['Num_No_Shows']/df_age['Total_Appts']

df_age 
```

```python
plt.bar(df_age.index, df_age['Prop_NoShows'])
plt.ylabel("Proportion")
plt.xlabel("Age Group")
plt.title("Proportion of No-Shows by Age Group")
plt.savefig("agegroup.jpg")
plt.show()
```

Thus, we can see that the age group with the highest proportion of no-shows is Group 2 (ages 19-37). The age group with the *least* no-shows is the oldest group, Group 4 (ages 55+). Let's next further divide this down by gender.

```python
#Create table containing # of total appts, # of no-shows, and proportion of no-shows for each age group & gender
df_gender = df.pivot_table(index = ["Age_Group", "Gender"], values = 'AppointmentID', aggfunc = 'count')
df_gender.rename(columns = {"AppointmentID": "Total_Appts"}, inplace = True)

df_gender['Num_No_Shows'] = df_noshow_ages.groupby(["Age_Group", "Gender"]).count()['AppointmentID']
df_gender["Prop_NoShows"] = df_gender['Num_No_Shows']/df_gender['Total_Appts']

df_gender
```

```python
test = df_gender.unstack().plot(y = "Prop_NoShows", kind='bar')
plt.ylabel("Proportion")
plt.xlabel("Age Group")
plt.title("Proportion of No-Shows by Age Group and Gender")
plt.savefig("gender.jpg")
plt.show()
```

We can compare this to the overall rates for each gender: 

```python
all_gender = df_gender.groupby('Gender')['Total_Appts', 'Num_No_Shows'].sum()
all_gender["Prop_NoShows"] = all_gender['Num_No_Shows']/all_gender['Total_Appts']
all_gender
```

From these results, it appears that men are, in general, more likely to be a no-show. Further discussion can be found in the Conclusions section. 

<a id='question3'></a>
### Research Question 3 : Is there a relationship between receiving an SMS reminder and showing up to an appointment? 

The column "SMS_received" indicates whether or not the patient received an SMS reminder message about the appointment. We will investigate whether or not there is a correlation betwen receiving a reminder and showing up to the appointment.  

```python
print("The number of patients who received a reminder is {}, out of {} total appointments".format(df.SMS_received.sum(), df.shape[0]))
print("The number of patients who recieved a reminder and did not show up is {}".format(df_noshow.SMS_received.sum()))
```

Putting this information into a table to make a pie chart: 

```python
table = df.pivot_table(index = ["SMS_received"], values = 'AppointmentID', columns = 'No-show', aggfunc = 'count')
table.rename(columns = {'No': 'Attended appt', 'Yes': 'Missed appt'}, index = {0: 'No SMS reminder', 1: 'Received SMS reminder'}, inplace = True)
table
```

```python
ax = table.plot(subplots = True, kind = 'pie', autopct='%1.1f%%', figsize = (10,5), title = ('Attended Appt', 'Missed Appt'), legend = None)
plt.suptitle('SMS Reminder Percentages', x = 0.5, y = 1.05, ha='center', fontsize='xx-large');
for x in ax: 
    x.set_ylabel('')
plt.show()
```

Let's take the transpose of the table to see the proportion of those who received or didn't receive reminders who attended their appointment. 

```python
ax2 = table.transpose().plot(subplots = True, kind = 'pie', autopct='%1.1f%%', figsize = (10,5), title = ('No SMS Reminder', 'Received SMS Reminder'), legend = None)
plt.suptitle('Missed Appointment Percentages', x = 0.5, y = 1.05, ha='center', fontsize='xx-large');
for x in ax2: 
    x.set_ylabel('')
plt.show()
```

## Conclusions

In this project, we examined three research questions. 

### Research Question 1: Neighborhoods

The first research question was a simple look at which neighborhoods had the highest and lowest rate of no-shows. The 10 neighborhoods with the highest rates of no-shows are: **Santos Dumont, Santa Cecília, Santa Clara, Itararé, Jesus de Nazareth, Horto, Ilha do Príncipe, Caratoíra, Andorinhas**, and **Praia do Suá**. On the other hand, the neighborhoods with the lowest rates of no-shows are: **Mata da Praia, Morada de Camburi, Vila Rubim, Do Quadro, Jardim da Penha, Santa Martha, Do Cabral, De Lourdes, Solon Borges**, and **Mário Cypreste**.

### Research Question 2: Age and Gender

The next research question sought to examine which age groups and gender had the highest rates of no-shows. From highest to lowest rates of no-shows, the age groups are: 
* **Group 2** (Ages 19 - 37): 23.5%
* **Group 1** (Ages 0 - 18): 22.0%
* **Group 3** (Ages 37 - 55): 19.5%
* **Group 4** (Ages 55+): 15.6%

When performing a further analysis on the percentage of no-shows by gender for each age group, it was found that women had a slightly higher percentage of no-shows across all four age groups. This was true overall as well: the no-show rate was 20.3% overall for women and 20.0% overall for men. 

These results do not have predictive power without further statistical analyses; these merely indicate historical trends in the data. 

### Research Question 3: SMS Reminders

Finally, we examined whether there was a correlation between receiving a SMS reminder and showing up to an appointment. 

Out of those who did NOT receive a SMS reminder, 83.3% attended their appointment, while 16.7% did not. Out of those who did receive a SMS reminder, 72.4% attended their appointment, while 27.6% did not. 

Out of those who attended their appointment, 70.9% had not received an SMS reminder, while 29.1% had not. Out of those who missed their appointment, 56.2% had received an SMS reminder, while 43.8% had not. 

Without further analyses, these results do not offer any definitive conclusions on the causal effects of SMS reminders. However, there does not appear to be a significant correlation betwen receiving an SMS reminder and showing up to an appointment. 



In sum, there does not seem to be a strong correlation between no-shows and any single characteristic from the dataset. However, this was merely a preliminary overview of the data-- no conclusions can be drawn without further statistical analyses. 

Future analysis should take a more granular view. For instance, it would be interesting to investigate whether certain days of the week had more no-shows than others. Additionally, we could also study whether receiving a SMS message was more useful for certain age groups than others.
